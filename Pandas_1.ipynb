### What is Pandas

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

```python
# string
country = ['Bangladesh','India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]

pd.Series(runs)
```

```python
#c Custom index

marks=[86,99,95,89,91]
subjects=['Bangla','Math','Physics','Chemistry','English']

pd.Series(marks,index=subjects)
```

```python
pd.Series(marks,index=subjects,name="Rafayel's mark")
```

```python
# series from dictionary

marks={

    'Bangla':85,
    'Math':95,
    'Physics':92,
    'Chemistry':89,
    'English':93
}
marks_series=pd.Series(marks,name="Rafayel's marks")
marks_series
```
